# Simple Visualization using **Matplotlib**

**Bar chart showing Attack category distribution**

```python
attack_counts = df['attack_cat'].value_counts()

plt.figure(figsize=(10,5))
plt.bar(attack_counts.index, attack_counts.values, color='skyblue')
plt.title("Attack Category Distribution")
plt.xlabel("Attack Category")
plt.ylabel("Number of Records")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Piechart showing ratio between normal and attack traffic.

```python
labels = ['Normal', 'Attack']
sizes = df['label'].value_counts().sort_index()
colors = ['lightgreen', 'salmon']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title("Normal vs Attack Traffic Ratio")
plt.show()
```

# Histogram showing Distribution of packet rate

```python
plt.figure(figsize=(8,5))
plt.hist(df['rate'], bins=40, color='steelblue', edgecolor='black')
plt.title("Distribution of Packet Rate")
plt.xlabel("Rate")
plt.ylabel("Frequency")
plt.show()
```

# Scatter plot visualizing relationship between connection duration and data size.

```python
plt.figure(figsize=(8,5))
plt.scatter(df['dur'], df['sbytes'], alpha=0.3, color='orange')
plt.title("Duration vs Source Bytes")
plt.xlabel("Duration")
plt.ylabel("Source Bytes")
plt.show()
```

# Line Chart Showing the mean traffic rate for each protocol type

```python
avg_rate = df.groupby('proto')['rate'].mean()

plt.figure(figsize=(8,5))
plt.plot(avg_rate.index, avg_rate.values, marker='o', color='purple')
plt.title("Average Packet Rate by Protocol")
plt.xlabel("Protocol")
plt.ylabel("Average Rate")
plt.grid(True)
plt.show()
```
